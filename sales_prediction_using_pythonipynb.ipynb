```python
###SELECTING LABEL AND TARGET FOR MODELING

x=data.drop(columns=["Sales","Newspaper"])
y=data[['Sales']]
```

```python
x.head()
```

```python
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2,random_state=42)
```

```python
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

```python
###CREATING LINEAR REGRESSION
```

```python
model= LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
####MODEL PREDICTION FOR TRAINING AND TESTING
```

```python
train_pred = model.predict(x_train)
test_pred = model.predict(x_test)
```

```python
####CHECKING ACCURACY OF MODEL PREDICTION
```

```python
print("Accuracy on Training Data is: ",r2_score(y_train,train_pred)*100)
print("Accuracy on Testing Data is: ",r2_score(y_test,test_pred)*100)
```

```python
mean_squared_error(y_test,test_pred)
```

```python
rmse = 0.5**mean_squared_error(y_test,test_pred)
rmse
```

```python
mean_absolute_error(y_test,test_pred)
```

```python
mean_absolute_percentage_error(y_test,test_pred)
```
